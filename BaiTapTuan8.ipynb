```python
#w7a17
def find_elm(lst):
    dic = {}
    for i in range(len(lst)):
        if lst[i] not in dic:
            dic[lst[i]] = 0
        dic[lst[i]] = dic[lst[i]] + 1
    if dic[lst[i]] > len(lst)/2:
        print(lst[i])
    else:
        print()
lst = list(map(int, input().split()))
find_elm(lst)
```

```python
#w7a18
def sap_xep(lst):
    lst1 = []
    lst2 = []
    for i in range(len(lst)):
        if lst[i] % 2 == 0:
            lst1.append(lst[i])
        else:
            lst2.append(lst[i])
    print(*(sorted(lst1) + sorted(lst2)[::-1]))
lst = list(map(int, input().split()))
sap_xep(lst)
```

```python
#w7a20
def sum_of_number(lst, k):
    a = 10**k
    b = 10**(k-1)
    lst1 = []
    for i in range(len(lst)):
        if lst[i] // a == 0 and lst[i] // b != 0:
            lst1.append(lst[i])
    print(sum(lst1), lst1)
lst = list(map(int, input().split()))
k = int(input())
sum_of_number(lst, k)

```

```python
#w7a21
def ngay_thang(N, T):
    tong_ngay = 0
    if T > 12:
        print('invalid')
    elif T <= 12:   
        for i in range(1, T):
            if i < 8:
                if i == 2:
                    tong_ngay = tong_ngay + 29
                elif i % 2 == 0:
                    tong_ngay = tong_ngay + 30
                else:
                    tong_ngay = tong_ngay + 31
            if i >=8:
                if i % 2 == 0:
                    tong_ngay = tong_ngay + 31
                else:
                    tong_ngay = tong_ngay + 30
        print(tong_ngay + N)
N, T = map(int, input().split())
ngay_thang(N, T)
```

```python
#w7a24
def delete_words(lst):
    lst1 = []
    for i in range(len(lst)):
        if lst[i] != lst[i-1]:
            lst1.append(lst[i])
    print(*(lst1))

lst = list(input().split())
delete_words(lst)
```

```python
#w7a25
def find_number_max(lst, k):
    stack = []
    remove = len(lst) - k
    
    for i in lst:
        while stack and remove > 0 and stack[-1] < i:
            stack.pop()
            remove = remove - 1
        stack.append(i)
    return ''.join(map(str, stack[:k]))
lst = list(map(int, input().strip()))
k = int(input())
print(find_number_max(lst, k))       


```
