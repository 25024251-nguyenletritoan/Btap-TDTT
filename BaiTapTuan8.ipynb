```python
#w7a16
def mang_dep(lst):
    dic = {}
    remove = 0
    for i in range(len(lst)):
        if lst[i] not in dic:
            dic[lst[i]] = 0
        dic[lst[i]] = dic[lst[i]] + 1

    for k, v in dic.items():
        if v > k:
            remove = remove + v - k
        elif v < k:
            remove = remove + v
    
    if remove == 0:
        print("YES")
    else: 
        print("NO", remove)
    
lst = list(map(int, input().split()))
mang_dep(lst)
```

```python
#w7a25
def find_number_max(lst, k):
    stack = []
    remove = len(lst) - k
    
    for i in lst:
        while stack and remove > 0 and stack[-1] < i:
            stack.pop()
            remove = remove - 1
        stack.append(i)
    return ''.join(map(str, stack[:k]))
lst = list(map(int, input().strip()))
k = int(input())
print(find_number_max(lst, k))       


```

```python
#w7a30
def find_so(lst):
    n = len(lst)
    daycon = [1] *n 
    father = [-1] *n 
    for i in range(n):
        for j in range(i):
            if lst[j] <= lst[i] and daycon[j] + 1 > daycon[i]:
                daycon[i] = daycon[j] + 1
                father[i] = j
    idx = daycon.index(max(daycon))
    lis = []
    while idx != -1:
        lis.append(lst[idx])
        idx = father[idx]
    lis.reverse()
    print(*(lis))
lst = list(map(int, input().split()))
find_so(lst)
```
