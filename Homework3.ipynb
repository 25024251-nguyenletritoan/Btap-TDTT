```python
#Bai 4
h = int(input())
for i in range (0, h):
    for j in range (h-i-1):
        print(" ", end="")
    if i == 0:
        print("*", end="")
    elif i == h-1:
        print("*" * (2*h-1), end="")
    else:
        print("*" + " "*(2*i-1) + "*", end="")
    print()
```
