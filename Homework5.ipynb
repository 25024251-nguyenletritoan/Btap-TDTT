```python
#w6a17
r, c = map(int, input().split())
matrix = []
for i in range(1, r+1):
    row = []
    for j in range(1, c+1):
        gtri = int(input(f"a[{i}][{j}]"))
        row.append(gtri)
    matrix.append(row)

lst_cheo_chinh = []
lst_cheo_phu = []
for i in range(1, r+1):
    for j in range(1, c+1):
        if i == j:
            lst_cheo_chinh.append(matrix[i-1][j-1])
        elif i + j == c +1:
            lst_cheo_phu.append(matrix[i-1][j-1])
print(*lst_cheo_chinh)
print(*lst_cheo_phu)

```

```python
#w6a18
n, m, k = map(int, input().split())
matrix = []
for i in range(1, n+1):
    row = []
    for j in range(1, m+1):
        gtri = int(input(f"a[{i}][{j}]"))
        row.append(gtri)
    matrix.append(row)

tong_hang = sum(matrix[k-1])
tong_cot = 0
for i in range(0, n):
        tong_cot = tong_cot + matrix[i][k-1]
tong_pt = tong_hang + tong_cot - matrix[k-1][k-1]
print(tong_pt)

```

```python
#w6a20
def find_word(lst: list, k: int):
    for i in range(len(lst)):
        if k == lst[i]:
            return i
    else:
        return -1

lst = list(map(int, input().split()))
k = int(input())
print(find_word(lst, k))
```

```python
#w6a21
chuoi = input().split()
counts = {}
for i in range(len(chuoi)):
    if chuoi[i] in counts:
        counts[chuoi[i]] = counts[chuoi[i]] + (i,)
    else:
        counts[chuoi[i]] = (i,)

print("{")
for k, v in counts.items():
    print(f"'{k}': {v},")
print("}")

```

```python
#w6a21 cach 2
def check_site(chuoi):
    tudien={}
    for i,tu in enumerate(chuoi):
        if tu in tudien:
            tudien[tu] = tudien[tu] + (i,)
        else:
            tudien[tu] = (i,)
    return tudien
chuoi = input().split()
print(check_site(chuoi))
```

```python
#w6a23
def de_number(lst: list, x: int):
    lst2= []
    for i in range(len(lst)):
        if lst[i] >= x:
            lst2.append(lst[i])
    return lst2

lst = list(map(int, input().split()))
x = float(input())
kq = de_number(lst, x)
print(*kq)
```

```python
#w6a24
def push_0(lst: list):
    so_0 = lst.count(0)
    con_lai = [x for x in lst if x!=0]
    a= [0]*so_0 + con_lai
    return a

lst = list(map(int, input().split()))
ket_qua = push_0(lst)
print(*ket_qua)
```

```python
#w6a25
def check_num(lst: list, k: int):
    return k not in lst

lst = list(map(int, input().split()))
for i in range(len(lst)):
    while lst[i] > 9 or lst[i] < 0:
        lst[i] = int(input())
k = int(input())
print(check_num(lst, k))
```

```python
#w6a27
lst = list(map(int, input().split()))
x = int(input())
so_be_hon = []
co_be_hon =False
for i in range(len(lst)):
    if lst[i] < x:
        so_be_hon.append(lst[i])
        co_be_hon = True

if co_be_hon:
    for x in so_be_hon:
        print(x, end= " ")
else:
    print("-1")
```
