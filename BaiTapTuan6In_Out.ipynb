```python
#w6a4
lst = list(input().split())
lst_k = []
lst_v = []
for i in lst:
    k, v = i.split(":")
    lst_k.append(k)
    lst_v.append(v)
dic = {}
for j in range(len(lst_k)):
    if lst_k[j] not in dic:
        dic[lst_k[j]] = []
    dic[lst_k[j]].append(lst_v[j])

print(dic)
```

```python
#w6a5
lst = list(map(int, input().split()))
dic = {'positives': 0, 'negatives': 0, 'zeros': 0}
for i in lst:
    if i > 0:
        dic['positives'] += 1
    elif i < 0:
        dic['negatives'] += 1
    else:
        dic['zeros'] += 1
print(dic)
```

```python
#w6a10
lst = list(map(int, input().split()))
k = int(input())
lst1 = []
for i in range(len(lst)):
    for j in range(i+1, len(lst)):
        if lst[i] + lst[j] == k and lst[i] != lst[j]:
            tup = (lst[i], lst[j])
            if tup not in lst1:
                lst1.append(tup)
print(lst1)
```

```python
#w6a12
lst = list(map(int, input().split()))
lst1 = []
lst2 = []
dic = {}
for i in range(len(lst)):
    if lst[i] not in dic:
        dic[lst[i]] = 0
    dic[lst[i]] += 1
for j in dic.values():
    lst1.append(j)
for h in dic.keys():
    lst2.append(h)

max_xh = lst1[0]
for g in lst1:
    if g > max_xh:
        max_xh = g

for i in range(len(lst2)):
    if dic[lst2[i]] == max_xh:
            kq = lst2[i]
            break
for i in range(len(lst2)):
     if dic[lst2[i]] == max_xh:
          if lst2[i] < kq:
               kq = lst2[i]
print(kq)
```
