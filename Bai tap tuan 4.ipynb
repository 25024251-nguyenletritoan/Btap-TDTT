```python
def dao_nguoc(n):
    return int(str(n)[::-1])  # Chuyển thành chuỗi -> đảo ngược -> chuyển lại thành số

# --- Chương trình chính ---
n = int(input("Nhập n: "))
buoc = 0

while n != dao_nguoc(n):      # Nếu n chưa đối xứng
    n = n + dao_nguoc(n)      # Lấy n cộng với số đảo của nó
    buoc += 1                 # Tăng số bước lên 1

print(f"{buoc} {n}")
```
