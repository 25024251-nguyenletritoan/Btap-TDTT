```python
#w7a1
def binary_search(arr, L, R, va):
    while L <= R:
        M = (L+R)//2
        if arr[M] == va:
            return M
        elif arr[M] < va:
            L = M + 1
        elif arr[M] > va:
            R = M - 1
    return -1
arr = list(map(int, input().split()))
va = int(input())
L = 0
R = len(arr) - 1
print(binary_search(arr, L, R, va))
```

```python
#w7a3 bubble sort
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr) - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
```

```python
#w7a3 selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        x = arr[i]
        m = min(arr[i::])
        a = arr.index(m, i)
        if m < x:
            arr[i], arr[a] =  arr[a], arr[i]
    return arr
arr = list(map(int, input().split()))
print(selection_sort(arr))

```

```python
#w7a7
def find_out(lst):
    dp = [1] * len(lst)
    for i in range(len(lst)):
        for j in range(i):
            if lst[i] > lst[j]:
                if dp[j] + 1 > dp[i]:
                    dp[i] = dp[j] + 1
    return max(dp)
lst = eval(input())
print(find_out(lst))
```

```python
#w7a8 không dùng binary search
def tim(lst1, lst2):
    ans = []
    for q in lst2:
        best = float('inf')
        for l, r in lst1:
            if l <= q <= r:
                length = r - l + 1
                if length < best:
                    best = length
        if best == float('inf'):
            return -1
        else:
            ans.append(best)
    return ans
lst1 = eval(input())
lst2 = eval(input())
print(tim(lst1, lst2))

```

```python
#w7a8 dùng binary search tham khảo mạng
def left_pos(arr, x):
    l, r = 0, len(arr)
    while l < r:
        m = (l+r)//2
        if arr[m][0] < x:
            l = m + 1
        else:
            r = m
    return l

def right_pos(arr, x):
    l, r = 0, len(arr)
    while l < r:
        m = (l + r) // 2
        if arr[m][0] <= x:
            l = m + 1
        else:
            r = m
    return l

def mininterval(intervals, queries):
    intervals.sort(key=lambda x: x[1] - x[0] + 1)
    qs = []
    for i in range(len(queries)):
        qs.append((queries[i], i))
    qs.sort()
    ans = [-1] * len(queries)
    
    for l, r in intervals:
        length = r - l + 1
        start = left_pos(qs, l)
        end = right_pos(qs, r)

        for i in range(start, end):
            idx = qs[i][1]
            if ans[idx] == -1:
                ans[idx] = length
    return ans
intervals = eval(input())
queries = eval(input())
print(mininterval(intervals, queries))
```

```python
#w7a13
def find_length_haihoa(lst, n):
    dem_max = 0
    for i in range(len(lst)):
        dem = 0
        for j in range(len(lst)):
            if lst[j] == lst[i]:
                dem = dem + 1   
            elif lst[j] - lst[i] == 1:
                dem = dem + 1
        if dem > dem_max:
            dem_max = dem
            
    print(dem_max)
lst = list(map(int, input().split()))
n = int(input())
while len(lst) != n:
    lst = list(map(int, input().split()))
find_length_haihoa(lst, n)
```

```python
#w7a13 dung dictionary
def find_length_haihoa(lst):
    dic = {}
    for k in range(len(lst)):
        if lst[k] not in dic:
            dic[lst[k]] = 0
        dic[lst[k]] = dic[lst[k]] + 1
    ans = 0
    for x in dic:
        if x + 1 in dic:
            current_length = dic[x] + dic[x+1]
            if current_length > ans:
                ans = current_length
    print(ans)
lst = list(map(int, input().split()))
find_length_haihoa(lst)
```
