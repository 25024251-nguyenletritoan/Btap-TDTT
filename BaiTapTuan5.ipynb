```python
#w5a2
def swap(a, b: int):
    return b, a
a, b = map(int, input().split())
print(swap(a, b))
```

```python
#w5a3
def so_nguyen_to(a: int):
    if a == 2:
        return True
    for i in range(2, a):
        nguyen_to = True
        if a % i == 0:
            nguyen_to = False
            break
        else:
            nguyen_to = True
    return nguyen_to

a = int(input())
print(so_nguyen_to(a))
```

```python
#w5a4
def perfect_number(a: int):
    tong = 0
    for i in range(1, a):
        if a % i == 0:
            tong = tong + i
    if tong == a:
        return True
    else:
        return False

a = int(input())
print(perfect_number(a))
```

```python
#w5a5
def check_site(a: list, b: int):
    for i in range(len(a)):
        if a[i] == b:
            return i
    else:
         return -1
        
a = list(map(int, input().split()))
b = int(input())
print(check_site(a, b))
        
```

```python
#w5a5 cach 2
def check_site(a: list, b: int):
    if b in a:
        return a.index(b)
    return -1

a = list(map(int, input().split()))
b = int(input())
print(check_site(a, b))


```

```python
#w5a6 
def giai_thua(n: int):
    tich = 1
    for i in range(1, n+1):
        tich = tich * i
    return tich

n = int(input())
print(giai_thua(n))
```

```python
#w5a7
def cal(a, b : float):
    return f"{(a-b):.2f}", f"{(a+b):.2f}", f"{(a*b):.2f}", f"{(a/b):.2f}"

a, b = map(float, input().split())
print(cal(a, b))
```

```python
#w5a8
n, m = map(int(input().split()))
cnt = 0
while n!=0 or m != 0:
    if n%2 != m%2:
        cnt = cnt + 1
    n = n >> 1
    m = m >> 1

print(cnt)
```

```python
#w5a9
def sum_num(a:int):
    a = abs(a)
    tong = 0
    while a > 0:
        tong = tong + (a%10)
        a = a // 10
    return tong

a = int(input())
print(sum_num(a))
```
