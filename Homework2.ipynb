```python
#Bai3
a=float(input("Canh dau tien cua tam giac="))
b=float(input("Nhap canh thu 2 cua tam giac la:"))
c=float(input("Nhap canh thu 3 cua tam giac la"))
if a+b<c or a+c<b or b+c<a:
    print("khong la ba canh cua tam giac")
elif a==b==c:
    C=a+b+c
    S=0.5*((3**(1/2))/2)*(a**2)
    print("Day la tam giac deu co chu vi la", C, "Va dien tich la", S)
elif a==b!=c or a==c!=b or b==c!=a:
    C=a+b+c
    S=((C/2)*(C-a)*(C-b)*(C-c))**(1/2)
    print("Day la tam giac can co chu vi la", C, "Va dien tich la", S)
elif (a**2)+(b**2)==(c**2) or (b**2)+(c**2)==(a**2) or (a**2)+(c**2)==(b**2):
    C=a+b+c
    S=((C/2)*(C-a)*(C-b)*(C-c))**(1/2)
    print("Day la tam giac vuong co chu vi la", C, "Va dien tich la", S)
else:
    C=a+b+c
    S=((C/2)*(C-a)*(C-b)*(C-c))**(1/2)
    print("Day la tam giac thuong co chu vi la", C, "Va dien tich la", S)

```

```python
#Bai5:
a=float(input("Nhap a:"))
b=float(input("Nhap b:"))
c=float(input("Nhap c:"))
delta=((b**2)-(4*a*c))
if delta>0:
    x1=(-b+(delta)**(0.5))/(2*a)
    x2=(-b-(delta)**(0.5))/(2*a)
    print("Phuong trinh co 2 nghiem phan biet la:", x1, "va", x2)
elif delta==0:
    x=(-b)/(2*a)
    print("Phuong trinh co nghiem duy nhat la", x)
elif delta<0:
    print("Phuong trinh vo nghiem")
```
